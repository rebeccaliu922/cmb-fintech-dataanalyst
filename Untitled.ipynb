```python
# Visualization continuous features
# age, job_year, frs_agn_dt_cnt, cur_debit_min_opn_dt_cnt, cur_credit_min_opn_dt_cnt
fig = plt.figure(figsize=(10, 20))
ax1=plt.subplot2grid((5,1),(0,0),)
ax2=plt.subplot2grid((5,1),(1,0),)
ax3=plt.subplot2grid((5,1),(2,0),)
ax4=plt.subplot2grid((5,1),(3,0),)
ax5=plt.subplot2grid((5,1),(4,0),)
sns.distplot(df1.age,ax=ax1)#年龄
sns.countplot(df1.job_year,ax=ax2)#工作年限
sns.distplot(df1.frs_agn_dt_cnt,ax=ax3)#首次代发工资距今天数
sns.distplot(df1.cur_debit_min_opn_dt_cnt,ax=ax4)#持有招行借记卡天数
sns.distplot(df1.cur_credit_min_opn_dt_cnt, ax=ax5)#持有招行信用卡天数
plt.tight_layout()
plt.suptitle('Exploratory analysis of raw data',y=1.02,fontsize=16,weight='bold')
```

```python
col=list(df1.columns)
catagorial_col=[]
for ii, col_name in enumerate(col):
    continuous_col=['age', 'job_year', 'frs_agn_dt_cnt', 'cur_debit_min_opn_dt_cnt', 'cur_credit_min_opn_dt_cnt']
    if col_name not in continuous_col:
        catagorial_col.append(col_name)
catagorial_col=catagorial_col[2:]
np.shape(catagorial_col)
```

```python
# Visualization catagorial data.
fig = plt.figure(figsize=(10, 180))
for ii in range(np.shape(catagorial_col)[0]):
    ax = plt.subplot2grid((np.shape(catagorial_col)[0], 1), (ii, 0), )
    sns.countplot(x=catagorial_col[ii], data=df1, ax=ax)
    
    
#ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
#ax2 = plt.subplot2grid((3, 3), (1, 0),)
#ax3 = plt.subplot2grid((3, 3), (1, 1),)
#ax4 = plt.subplot2grid((3, 3), (1, 2),)
#ax5 = plt.subplot2grid((3, 3), (2, 0), colspan=3)
#for col, ax in zip(plotlist, fig.get_axes()):
#    sns.countplot(x=col, data=df, ax=ax)
plt.suptitle('Exploratory analysis of cleaned data', y=1.01, fontsize=16, weight='bold')
plt.tight_layout()
```
